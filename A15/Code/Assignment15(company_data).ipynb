```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn import tree
```

```python
df = pd.read_csv("Company_Data.csv")
df
```

# **EDA**


```python
#Missing values
df.isnull().sum()
```

```python
df.describe()
```

```python
company_val = []
for value in df["Sales"]:
     if value<=7.49:
        company_val.append("low")
     else:
        company_val.append("high")

df["company_val"]= company_val
df
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["ShelveLoc"] = le.fit_transform(df["ShelveLoc"])
df["Urban"] = le.fit_transform(df["Urban"])
df["US"] = le.fit_transform(df["US"])
df["company_val"] = le.fit_transform(df["company_val"])
df
```

```python
x = df.drop(['company_val','Sales'], axis =1)
y = df['company_val']
```

# **Ensembling technique-Random forest**

```python
#random forest

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(criterion = "entropy",n_estimators=300,max_features=3)
```

```python
from sklearn.model_selection import cross_val_score
res3 = cross_val_score(rf,x,y,cv = 10)
```

```python
print(res3.mean())
```

# **Bagging**

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=10,random_state=7,shuffle=True)
```

```python
from sklearn.linear_model import LogisticRegression
m1 = LogisticRegression()
```

```python
from sklearn.ensemble import BaggingClassifier

model1 = BaggingClassifier(base_estimator= m1, n_estimators=20,random_state=7)
```

```python
res = cross_val_score(model1,x,y,cv = kf)
```

```python
print(res)
```

```python
print(res.mean())
```

# **Stacking**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
```

```python
estimators = []
```

```python
m1 = LogisticRegression(max_iter = 500)
estimators.append(("logistic",m1))
```

```python
m2 = DecisionTreeClassifier(criterion ="gini",max_depth = 3)
estimators.append(("CART",m2))
```

```python
m3 = SVC()
estimators.append(("SVM",m3))
```

```python
ens_model = VotingClassifier(estimators)
```

```python
res = cross_val_score(ens_model,x,y,cv =10)
```

```python
print(res.mean())
```
