```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import norm
```

```python
# Imprting data
data = pd.read_csv("Cutlets.csv")
data.head()
```

```python
#H1 = mean of unitA not equal to unitB
#H0 = mean of unitA equal to unitB

unitA = pd.Series(data.iloc[:,0])
unitA
```

```python
unitB = pd.Series(data.iloc[:,1])
unitB
```

```python
#Using 2 sample 2 tail ttest
stats.ttest_ind(unitA,unitB)
```

```python
#pvalue is 0.472 it is greater than 0.05 (Î±,significance level) so null hypothesis(H0) is true
# That is mean of unitA is equal to unitB
```

## **Question 2**

```python
#H0 = Average turn around time of reports of the laboratories are same
#H1 = One of the average turn around time of reports of the laboratories are not same
```

```python
#Read data
data1 = pd.read_csv("LabTAT.csv")
data1.head()
```

```python
#Anova testing is used.Because comparing mean of more than two sample
stats.f_oneway(data1["Laboratory 1"],data1["Laboratory 2"],data1["Laboratory 3"],data1["Laboratory 4"])
```

```python
#p_value is very less than significance value 0.05,so
#Alternative hypothesis accepted,Difference in average turn around time of reports of the laboratories
```

# **Question 3**

```python
#H0 = All proportion are equal
#H1 = Not all proportion are equal
```

```python
d3 =pd.read_csv("BuyerRatio.csv")
d3.head()
```

```python
d3 = d3.drop("Observed Values",axis = 1)
d3
```

```python
stats.chi2_contingency(d3)
```

```python
#p_value is greater than significance level 0.05,so Null hypothesis is accepted.
# That is  All proportion are equal
```

```python
#Anova testing is used.
stats.f_oneway(d3["East"],d3["West"],d3["North"],d3["South"])
```

```python
#p_value is greater than significance level 0.05,so Null hypothesis is accepted.
# That is  All proportion are equal
```
