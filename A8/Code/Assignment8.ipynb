```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("wine.csv")
df
```

# **EDA**

```python
wine = df.copy()
```

```python
wine.info()
```

```python
wine.isna().sum()
```

```python
wine['Type'].value_counts()
```

```python
wine=wine.iloc[:,1:]
wine
```

```python
wine.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
wine_std = sc.fit_transform(wine)
wine_std
```

# **PCA Implimentation**

```python
from sklearn.decomposition import PCA
```

```python
# Applying PCA Fit Transform to dataset
pca = PCA(n_components=13)
```

```python
pca.fit_transform(wine_std)
```

```python
# PCA Components matrix or covariance Matrix
pca.components_
```

```python
#The amount of variance that each PCA has
var = pca.explained_variance_ratio_
var
```

```python
#cumulative variance
var1 = np.cumsum(np.round(var,4)*100)
var1
```

```python
plt.plot(var1)
plt.show()
```

```python
# Final Dataframe
final_df=pd.concat([df['Type'],pd.DataFrame(wine_std[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
sns.scatterplot(data=final_df,x="PC1",y="PC2",hue="Type")
```

```python
# Visualization of PCAs
fig=plt.figure(figsize=(10,10))
sns.scatterplot(data=final_df)
```

```python
final_wine=pd.concat([pd.DataFrame(wine_std[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_wine
```

# Checking with other Clustering Algorithms

# **Heirarchical clusters using pca**





```python
h_cluster = final_wine
```

```python
# create dendrogram
import scipy.cluster.hierarchy as sch
sch.dendrogram(sch.linkage(h_cluster, method='complete'))
plt.show()
```

```python
# create clusters
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'complete')
hc.fit(h_cluster)
```

```python
yhc = hc.fit_predict(h_cluster)
yhc
```

```python
wine1 = df.copy()
wine1["h_cluster"] = yhc
wine1
```

```python
wine1['h_cluster'].value_counts()
```

```python
# Plotting barplot using groupby method to get visualization of how many row no. in each cluster
fig, ax = plt.subplots(figsize=(10, 6))
wine1.groupby(['h_cluster']).count()['Type'].plot(kind='bar')
plt.ylabel('ID Counts')
plt.title('Hierarchical Clustering',fontsize='large',fontweight='bold')
ax.set_xlabel('Clusters', fontsize='large', fontweight='bold')
ax.set_ylabel('ID counts', fontsize='large', fontweight='bold')
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.show()
```

# **K- means cluster using pca**

```python
k_cluster = final_wine
```

```python
#find the optimum number of clusters

from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(k_cluster)
  wcss.append(kmeans.inertia_)
```

```python
sns.lineplot(x=range(1,11),y=wcss)
```

```python
# Consider cluster = 3
kmeans = KMeans(n_clusters=3)
predict=kmeans.fit_predict(k_cluster.iloc[:,:])
predict
```

```python
wine1["k_cluster"] =kmeans.labels_
wine1
```

```python
wine1['k_cluster'].value_counts()
```
