```python
import pandas as pd
import numpy as np
```

```python
book_df = pd.read_csv('book.csv',encoding='latin-1')
book_df.head()
```

```python
#Rename columns
book_df = book_df.rename(columns = {"User.ID":"UserId","Unnamed: 0":"index","Book.Title":"Title","Book.Rating":"Rating"})
```

```python
book_df
```

```python
book_df['Title'].value_counts() # unique book count
```

```python
book_df.sort_values('UserId')
book_df.shape
```

```python
#number of unique users in the dataset
len(book_df.UserId.unique())
```

```python
#remove duplicates
book_df.drop_duplicates(subset=['UserId'],inplace=True)
```

```python
user_book_df = book_df.pivot(index='UserId',columns='Title',values='Rating')
user_book_df
```

```python
#Impute those NaNs with 0 values
user_book_df.fillna(0, inplace=True)
user_book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_book_df.values,metric='cosine')
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids
user_sim_df.index = book_df.UserId.unique()
user_sim_df.columns = book_df.UserId.unique()
user_sim_df
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:50]
```
